##PIVOTING##

```python
#what is pivot table in pandas?
#A pivot table in pandas is a tool to summarize and analyze the data
import pandas as pd
```

```python
data = {
    'student': ['a','b','c','a','b','c'],
    'subject': ['maths','maths','maths','science','science','science'],
    'marks':[85,90,95,80,70,88],
    'sem':['sem1','sem1','sem1','sem2','sem2','sem2']
}
df = pd.DataFrame(data)
df
```

```python
#synatx:- pd.pivot_table(data,values,index,columns,aggfunc)
pd.pivot_table(df,values='marks',index='student',columns='subject',aggfunc='mean')
```

```python
#add extra rows about student abc in above df
data2 = {
    'student' : ['a','b','c'],
    'subject':['maths','science','Histroy'],
    'marks':[82,91,90]}
df2 = pd.DataFrame(data2)
df = pd.concat([df,df2],ignore_index=True)
df
```

```python
pd.pivot_table(df,values='marks',index='student',columns='subject',aggfunc='mean')
```
